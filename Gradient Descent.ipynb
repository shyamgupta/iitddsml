# Gradient Descent

```python
import sympy
from sympy import *
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

# Cost Function

```python
x,y = symbols('x,y')
f = 85 - (1/90)*(x**2)*(x-6)*(y**2)*(y-6)
# We can see that the minimum is somewhere around (4,4)
sympy.plotting.plot3d(f,(x,0,5),(y,0,5),size=(10,7))
```

# Derivatives and Gradient

```python
def gradient_descent(f, iteration,learning_rate, x0, y0):
    for i in range(iteration):
        x0 = x0 - learning_rate*diff(f,x).evalf(subs={x:x0,y:y0})
        y0 = y0 - learning_rate*diff(f,y).evalf(subs={x:x0,y:y0})
    return x0,y0
```

```python
gradient_descent(f,250,0.05,0.5,0.6)
```
