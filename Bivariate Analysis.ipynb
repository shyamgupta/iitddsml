# Bivariate Analysis
- Uncovers insights embedded within two variables of interest: we are typically interested in how these two variables are distributed or related.
- We can have three possible combinations of variables namely: numerical-numerical, numerical-categorical, and categorical-categorical.

```python
import pandas as pd
from  matplotlib import pyplot as plt
import seaborn as sns
```

# Scatter Plot
Pattern on a scatter plot can provide insights into the strength and direction of the relationship between the two variables. This could either be positive (i.e., as one variable increases, the other variable increases) or negative (i.e., as one variable increases, the other variable decreases)
Data: https://github.com/PacktPublishing/Exploratory-Data-Analysis-with-Python-Cookbook/blob/main/Ch5/Data/penguins_size.csv

```python
df = pd.read_csv('data/penguins_size.csv')
df = df[['species','culmen_length_mm','body_mass_g']]
df.head(5)
```

```python
df.shape
```

```python
df.dtypes
```

```python
ax = sns.scatterplot(data=df, x = df['culmen_length_mm'],y=df['body_mass_g'])
ax.set_title('Bivariate Annalysis of Culmen Length and Body Mass of penguins')
```

# Crosstab / Tow-Way Table
- Displays the relationship between categorical variables in a matrix format.
- Rows = categories within the first categorical variable
- Columns = categories of the second categorical variable.
- Values =  Either the frequency of occurrence or the percentage of occurrence

```python
df = pd.read_csv('data/penguins_size.csv')
df = df[['species','culmen_length_mm','sex']]
df.head(5)
```

```python
pd.crosstab(index=df['species'],columns=df['sex'])
```

# Pairplots
- Pairplot is a visual representation of the distribution of single variables and the relationship between two variables.
- It is a group of subplots that display single variables (numerical or categorical) and two variables (numerical-numerical and categorical-numerical)
- Provides a simple way to view the distribution and relationship across multiple variables in our dataset.

```python
df = pd.read_csv('data/penguins_size.csv')
df = df[['species','culmen_length_mm','body_mass_g','sex']]
df.head(5)
```

```python
sns.pairplot(df)
```
