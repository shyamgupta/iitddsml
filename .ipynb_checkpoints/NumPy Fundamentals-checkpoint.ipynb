# What is NumPy?
Numpy facilitates efficient numerical operations on large quantities of data. It is designed for efficiency on large datasets.
1. NumPy internally stores data in a contiguous block of memory, independent of other built-in Python objects.
2. Numpy can operate on this memory without any type checking or other overhead.
3. NumPy arrays also use much less memory than built-in Python sequences.
4. Allows you to perform complex calculations on entire array without using for loops.
5. **Vectorization**: Any arithmetic operations between equal-size NumPy arrays apply the operation element-wise

# ndarray
NumPy provides a high-performance N-dimensional array object. It is a **multidimensional homogeneous** array with:
- Predetermined number of items: once you define their size at the time of creation, it remains unchanged
- All the items in it are of the same type (defined by `dtype` object) and the same size

**Shape**: Number of dimensions, it is a tuple of N positive integers that specify the size for each dimension.

**Axes** : In a 2-Dimensional NumPy array for example, Axes are the directions along the rows and columns. Axes are numbered, starting with 0.
- **axis 0** is the "first" axis that runs down the rows in a multi-dimensional array.
- **axis 1** is the "second" axis that runs across the columns in a multi-dimensional array.

**Rank** is the number of axes

# How do axes work?
Here we will see some examples of how NumPy axes work. Functions like `sum(), mean(), min(), median()` etc. **aggregate** your data. For example, if I add 5 numbers, the result will be a single number, i.e. a large number of values collapse into a single value.


```python
# Import NumPy
import numpy as np

# Create a 2D array
np_array_2d = np.arange(0,6).reshape(2,3)
np_array_2d
```

```python
# When we use the sum function with axis=0
np.sum(np_array_2d, axis=0)
```

```python
# # When we use the sum function with axis=1
np.sum(np_array_2d, axis=1)
```

Lets take anothere example of `np.concatenate()` function: here, the `axis` parameter determines the axis along which we stack the arrays

```python
nparray1 = np.array([[1,1,1],[1,1,1]])
nparray2 = np.array([[9,9,9],[9,9,9]])
print(nparray1)
print('\n')
print(nparray2)
```

```python
# Concatenate along axis=0
np.concatenate([nparray1,nparray2], axis=0)
```

```python
# Concatenate along axis=1
np.concatenate([nparray1,nparray2], axis=1)
```

 > # Axes and One-Dimensional NumPy Arrays:
> 1 dimensional NumPy arrays only have one axis, axis 0 and unlike 2D arrays it does NOT point downwards. `np.concatenate()` with `axis-0` will concatenate **horizontally**. Trying to contacetanat along `axis=1` will cause an `axis 1 out of bounds` error.

# Creating NumPy Arrays
In NumPy, brackets are used to identify the number of dimensions. When we pass a single set of `[]` to `np.array()`, we are crating a one-dimensional NumPy array, When we pass two sets of `[]` to `np.array()`, we are crating a two-dimensional NumPy array and so on

```python
# Pass a list of sequence of Python lists to np.array()
# It can also accept tuples and sequence of tuples
array1 = np.array([
    [1,2,3],
    [4,5,6]
    ])

print(f'The 2D array is: \n {array1}')
print(f'Type of object is {type(array1)}')
print(f'Data type of array is {array1.dtype}')
print(f'Number of dimensions is {array1.ndim}')
print(f'Size of the array is {array1.size}, which denotes number of elements')
print(f'Shape of the array is {array1.shape}')
print(f'Size in bytes of each item of the array is {array1.itemsize}')
```

# dtype
By default, `np.array()`  can associate the most suitable type according to the values contained in the NumPy array. We can also however, explicitly define the dtype using the `dtype` option as argument of the function.

```python
array1 = np.array([[1,2,3],[4,5,6]], dtype='float32')
```
