# Grouping Data

```python
# Get average (mean) number of purchases based on number of kids at home
# We are grouping by "Kidhome" column, and aggregating by "NumStorePurchases" column
df.groupby('Kidhome')['NumStorePurchases'].mean()
```

## `pd_dataframe.groupby()` Function
`pd_dataframe.groupby()` will split the data into groups based on predefined criteria. You can also group by more than one column by providing the list of column names to `groupby`. The output of `groupby` is  `GroupBy` object. We can use the following methods on this object:
1. `group_by_object.indices`: Shows how the groups were split. Output is a dictionary, where keys are the group names and values are indices of each group
2. `group_by_object.ngroups` will show the number of groups
3. `group_by_object.get_group('group_name')` to select a single group

# Merging Data
**We need a common column (join key column) across datasets on which we can perform a merge**.  The columns do not need to have the same name; they only need to have matching values within the two datasets.
1. **Left Join**: retain all the rows in the left dataframe. Values in the right dataframe that do not match the values in the left dataframe are added as `NaN`
2. **Right Join**: We retain all the rows in the right dataframe. Values in the left dataframe that do not match the values in the right dataframe are added as `NaN`
3. **Inner Join**: Retain only the common values in both the left and right dataframes in the result, there are no `NaN` values in this case
4. **Outer Join**: We retain all the rows for the left and right dataframes. If the values do not match, `NaN` is added to the result.


```python
df1 = pd.read_csv("data/marketing_campaign_merge1.csv")
df2 = pd.read_csv("data/marketing_campaign_merge2.csv")
```

```python
df1.head(5)
```

```python
df2.head(5)
```

```python
# Perform merge on ID column
merged_df = pd.merge(df1,df2, on="ID")
merged_df.head(5)
```

# Categorizing Data
Categorizing means binning, bucketing, or cutting a dataset. **It involves grouping the numeric values in a dataset into smaller intervals called bins or buckets**. When we bin numerical values, each bin becomes a categorical value.

```python
df.head(5)
```

```python
# Categorize the number of store purchases into high, moderate, and low categories
# Our bins can be interpreted as 0–4 (low), 5–8 (moderate), and 9–13 (high).
df['bins'] = pd.cut(x=df['NumStorePurchases'], bins=[0,4,8,13],labels=['Low','Moderate','High'])
df[['NumStorePurchases','bins']].head(5)
```

# Removing Duplicates

To the `drop_duplicates()` method, we supply the list of the subset columns as the first argument, we can drop all rows that contain duplicates based on those subset columns. This is useful when we have several columns and only a few key columns contain duplicate information. It allows us to keep instances of duplicates, using the keep parameter. With the keep parameter, we can specify whether we want to keep the “first” or “last” instance or drop all instances of the duplicate information. By default, the method keeps the first instance.

```python
df3 = pd.read_csv("data/marketing_campaign.csv")
df3 = df3[['Education','Marital_Status','Kidhome', 'Teenhome']]
df3.head(5)
```

```python
df3.shape
```

```python
df3_drop_duplicates = df3.drop_duplicates()
df3_drop_duplicates.shape
```
