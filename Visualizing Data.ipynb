```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('data/HousingPricesData.csv')
df = df[['Zip', 'Price', 'Area', 'Room']]
df.head(5)
```

```python
# Data has 924 rows and 4 columns
df.shape
```

```python
df.dtypes
```

```python
# Create a PriceperSqm variable based on the Price and Area variables:
df['PriceperSqm'] = df['Price']/df['Area']
df.head(5)
```

```python
# Sort house prices in descending order
df_sorted_price = df.sort_values('Price',ascending=False)
df_sorted_price.head(5)
```

# Bar Chart

```python
# subplots returns a tupple that has figure and axes objects
# with no arguments, subplots() will create 1 subplot
# figure: elements like legend, title, etc. Container for what we see on the page
# axes: Holds the data, type of graph etc.
fig,ax = plt.subplots()
ax.bar(df_sorted_price['Zip'][0:10],df_sorted_price['Price'][0:10])
ax.set_xlabel('Zip Code')
ax.set_ylabel('House price in millions')
ax.set_title('Top 10 House prices')
plt.show()
```

```python
# subplots(m,n) will create a grid wth m rows and n columns, where we can create subplots
# we are creating a grid with 2 rows and 1 column
fig,ax = plt.subplots(2,1,figsize=(12,6))
ax[0].bar(df_sorted_price['Zip'][0:10],df_sorted_price['Price'][0:10])
ax[1].bar(df_sorted_price['Zip'][0:10],df_sorted_price['PriceperSqm'][0:10])
ax[0].set_ylabel('House price in millions')
ax[1].set_ylabel('House price per sq. meter')
ax[1].set_xlabel('Zip Codes')
plt.show()
```
